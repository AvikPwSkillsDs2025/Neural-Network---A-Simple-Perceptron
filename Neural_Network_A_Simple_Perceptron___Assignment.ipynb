QUESTION 1:
What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

ANSWER:
Deep Learning is a subset of machine learning that uses neural networks with multiple layers to learn complex patterns from large datasets.

Evolution:
Early AI used rule-based systems. Later machine learning introduced algorithms like Decision Trees and SVM that required manual feature engineering. With the availability of large datasets and powerful GPUs, deep neural networks became practical. Deep learning automatically learns hierarchical features from raw data.

Difference from traditional ML:
Traditional machine learning requires manual feature extraction and works well on small datasets. Deep learning automatically learns features from raw data and performs better on large datasets such as images, audio, and text.

QUESTION 2:
Explain the basic architecture and functioning of a Perceptron. What are its limitations?

ANSWER:
A perceptron is the simplest neural network model used for binary classification.

Architecture:
Input layer receives features.
Weights multiply each input.
A weighted sum is computed.
An activation function produces the final output.

Functioning:
The perceptron calculates a weighted sum of inputs and applies an activation function to decide the class.

Limitations:
It can only solve linearly separable problems.
It cannot solve problems like XOR.
It has limited representational power compared to multilayer neural networks.

QUESTION 3:
Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

ANSWER:
Activation functions introduce non-linearity into neural networks so they can learn complex patterns.

Sigmoid:
Output range 0 to 1.
Used in binary classification.
Problem: vanishing gradient.

Tanh:
Output range -1 to 1.
Centered around zero which helps optimization.
Still suffers from vanishing gradient.

ReLU:
Output is max(0,x).
Very fast and widely used in deep networks.
Problem: dying ReLU for negative values.

QUESTION 4:
What is the difference between Loss function and Cost function in neural networks? Provide examples.

ANSWER:
Loss function measures the error for a single training example.

Cost function is the average of loss values across the entire dataset.

Examples:
Binary Cross Entropy for binary classification.
Mean Squared Error for regression.
Categorical Cross Entropy for multi-class classification.

QUESTION 5:
What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

ANSWER:
Optimizers adjust model weights to minimize the loss function during training.

Gradient Descent:
Updates weights using the gradient of the loss.
Simple but slow.

RMSprop:
Uses adaptive learning rate for each parameter.
Works well for non-stationary problems.

Adam:
Combines momentum and RMSprop.
Fast convergence and widely used in deep learning.

QUESTION 6:
Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

ANSWER:

```python
import numpy as np

X = np.array([
[0,0],
[0,1],
[1,0],
[1,1]
])

y = np.array([0,0,0,1])

weights = np.random.rand(2)
bias = np.random.rand()

lr = 0.1

for epoch in range(100):
    for i in range(len(X)):
        z = np.dot(X[i],weights) + bias
        pred = 1 if z > 0 else 0
        error = y[i] - pred
        weights += lr * error * X[i]
        bias += lr * error

for i in X:
    z = np.dot(i,weights) + bias
    pred = 1 if z > 0 else 0
    print(i,pred)
```

QUESTION 7:
Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

ANSWER:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)
tanh = np.tanh(x)

plt.plot(x,sigmoid,label="Sigmoid")
plt.plot(x,relu,label="ReLU")
plt.plot(x,tanh,label="Tanh")

plt.legend()
plt.show()
```

QUESTION 8:
Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

ANSWER:

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train,y_train),(X_test,y_test) = mnist.load_data()

X_train = X_train/255.0
X_test = X_test/255.0

model = Sequential([
Flatten(input_shape=(28,28)),
Dense(128,activation="relu"),
Dense(10,activation="softmax")
])

model.compile(optimizer="adam",
loss="sparse_categorical_crossentropy",
metrics=["accuracy"])

history = model.fit(X_train,y_train,epochs=5)

print("Training Accuracy:",history.history["accuracy"][-1])
```


QUESTION 9:
Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

ANSWER:

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten

(X_train,y_train),(X_test,y_test) = fashion_mnist.load_data()

X_train = X_train/255.0
X_test = X_test/255.0

model = Sequential([
Flatten(input_shape=(28,28)),
Dense(128,activation="relu"),
Dense(10,activation="softmax")
])

model.compile(optimizer="adam",
loss="sparse_categorical_crossentropy",
metrics=["accuracy"])

history = model.fit(X_train,y_train,epochs=5)

plt.plot(history.history["loss"],label="Loss")
plt.plot(history.history["accuracy"],label="Accuracy")
plt.legend()
plt.show()
```

QUESTION 10:
Explain the workflow for fraud detection using deep learning.

ANSWER:
Model Design:
Use a multilayer neural network with dense layers to learn complex transaction patterns.

Activation Function:
ReLU for hidden layers and Sigmoid for output since this is binary classification.

Loss Function:
Binary Cross Entropy because the task is fraud vs legitimate classification.

Handling Imbalance:
Use class weights or oversampling techniques like SMOTE.

Optimizer:
Adam optimizer for faster convergence.

Prevent Overfitting:
Use dropout layers, early stopping, and regularization.

Evaluation:
Use Precision, Recall, F1-score, and ROC-AUC instead of accuracy because fraud datasets are highly imbalanced.
